```python
import matplotlib.pyplot as plt
import numpy as np
import random
import math
```

```python
x=np.linspace(0,2*math.pi*3
              ,100)

y=np.sin(x)
```

```python

import subprocess
import matplotlib.pyplot as plt

def get_git_hash(short=True):
    try:
        cmd = ["git", "rev-parse", "--short", "HEAD"] if short else ["git", "rev-parse", "HEAD"]
        return subprocess.check_output(cmd, stderr=subprocess.DEVNULL).decode("utf-8").strip()
    except Exception:
        return "no-git"

```

```python

import subprocess

remote = subprocess.check_output(
    ["git", "config", "--get", "remote.origin.url"]
).decode().strip()

# For your example: git@github.com:jonnyhtw/scripts.git
slug = remote.split(":")[-1].replace(".git", "")
print(slug)  # Output: jonnyhtw/scripts
```

```python
fig, ax =plt.subplots(layout='constrained')

labels = ['model '+r'$\alpha$','model '+r'$\beta$']

markers=['o','s']

for i in range(2):
    ax.plot(x,y+np.random.uniform(-0.3, .3, 100), alpha=0.5, label=labels[i], marker=markers[i])



# ax.plot(x,y,'k--',label='Ground\ntruth')
# ax.fill_between(x,y-np.ones(len(y)),y+np.ones(len(y)),'k--',label='Ground\ntruth')
ax.fill_between(x,y-np.ones(len(y)),y+np.ones(len(y)), alpha=.25,label='Ground\ntruth')



ax.set_title('Temperature validation illustration')

# Compose label
commit = get_git_hash(short=True)
label = f"git: {commit}"



ax.legend(loc='upper right')

ax.set_xticks([])
ax.set_yticks([])

ax.set_facecolor('whitesmoke')

# Put in bottom-right corner of the *figure*
fig.text(
    0.05, 0.015, slug+'/'+label,
    ha="left", va="bottom",
    fontsize=10, color="#444",transform=ax.transAxes,
    family="monospace", alpha=0.8
)

# ax.set_ylim(bottom=-1.5)

plt.show();
```
