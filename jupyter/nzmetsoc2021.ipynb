```python
# https://stackoverflow.com/a/10138308/6288682

import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math


fig = plt.figure(figsize = [15,10])

mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 10000)
y = stats.norm.pdf(x, mu, sigma)
plt.plot(x, y)

indices = np.where((x > -1 )& (x < 1))
plt.fill_between(x[indices], y[indices], 0)

indices = np.where((x > -2 )& (x < -1))
plt.fill_between(x[indices], y[indices], 0, color='r', alpha = 0.5,) 

indices = np.where((x > 1 )& (x < 2))
plt.fill_between(x[indices], y[indices], 0, alpha = 0.5, color = 'r')

plt.ylim(bottom = 0)

fontsize = 25
plt.xticks(fontsize=fontsize)
plt.xlabel(r'$\sigma$', fontsize = fontsize)
ax = plt.gca()
ax.axes.yaxis.set_visible(False)

plt.grid()


plt.savefig('nzmetsoc2021.png')

plt.show()
```
