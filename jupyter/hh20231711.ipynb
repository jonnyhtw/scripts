```python
import iris
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import matplotlib.path as mpath
import numpy as np
```

# Open access 'HadISST' data

```python
cube = iris.load_cube('/nesi/project/niwa00013/williamsjh/NZESM/nzesm-historical-warming/HadISST_sst_1950_mdi.nc',
                      'sea_surface_temperature')
```

```python
cube[0]# use first month of the 12 monthly means from 1950 data
```

```python
latitude = cube.coord('latitude').points
```

```python
longitude = cube.coord('longitude').points
```

# projection info from ---> https://scitools.org.uk/cartopy/docs/v0.15/crs/projections.html

```python
fig = plt.figure(figsize=[5,5])

ax = plt.subplot(1,1,1, projection = ccrs.PlateCarree())

mesh = plt.pcolormesh(longitude, latitude, cube[0].data, transform = ccrs.PlateCarree())

plt.show();
```

```python
fig = plt.figure(figsize=[5,5])

ax = plt.subplot(1,1,1, projection = ccrs.Mollweide())

mesh = plt.pcolormesh(longitude, latitude, cube[0].data, transform = ccrs.PlateCarree())

plt.show();
```

```python
fig = plt.figure(figsize=[5,5])

ax = plt.subplot(1,1,1, projection = ccrs.InterruptedGoodeHomolosine())

mesh = plt.pcolormesh(longitude, latitude, cube[0].data, transform = ccrs.PlateCarree())

plt.show();
```

```python
fig = plt.figure(figsize=[5,5])

ax = plt.subplot(1,1,1, projection = ccrs.SouthPolarStereo())

mesh = plt.pcolormesh(longitude, latitude, cube[0].data, transform = ccrs.PlateCarree())

ax.set_extent([-180, 180, -90, -60], ccrs.PlateCarree())

plt.show();
```

# 'always circular example' from here ---> https://scitools.org.uk/cartopy/docs/v0.15/examples/always_circular_stereo.html


```python

fig = plt.figure(figsize = [10,5])

ax1 = plt.subplot(1,2,1,projection = ccrs.SouthPolarStereo())
ax2 = plt.subplot(1,2,2,projection = ccrs.SouthPolarStereo())

ax1.set_extent([-180,180,-90,-60], ccrs.PlateCarree())
ax2.set_extent([-180,180,-90,-60], ccrs.PlateCarree())

ax1.pcolormesh(longitude, latitude, cube[0].data, transform = ccrs.PlateCarree(), vmin=-1.8, vmax =5)

ax2.pcolormesh(longitude, latitude, cube[0].data, transform = ccrs.PlateCarree(), vmin=-1.8, vmax =5)

theta = np.linspace(0, 2*np.pi, 100)
center, radius = [0.5, 0.5], 0.5
verts = np.vstack([np.sin(theta), np.cos(theta)]).T
circle = mpath.Path(verts * radius + center)

ax2.set_boundary(circle, transform=ax2.transAxes)

ax2.coastlines()

ax2.gridlines()
ax1.grid()
```
